```python
from parse_log import parse
import pandas as pd
import numpy as np
from IPython.display import HTML, display
from bokeh.layouts import gridplot
from bokeh.palettes import Spectral11, Spectral4
from bokeh.plotting import figure, output_notebook, show

tables = parse('log/training_gail_trpo_noise_20171202.log')
display(HTML('''<b>The following tables are available:</b>
                <ul>%s</ul>''' % (''.join(sorted('<li style="float:left;margin-right:28px">%s</li>' % key for key in tables)))))

p1 = figure(plot_width=800, plot_height=450)
p1.title.text = 'Click on legend entries to hide the corresponding lines'

exlude = ['TimeElapsed', 'TimestepsSoFar', 'EpisodesSoFar', 'optimgain', 'entloss', 'entrpoy_loss']
zero_ones = []
added = 0
for i, (key, values) in enumerate(tables.items()):
    if key in exlude:
        continue
    if np.max(values) <= 1.001:
        zero_ones.append((key, values))
        continue
    p1.line(range(len(values)), values, line_width=2, alpha=0.8, legend=key, color=Spectral4[added%11])
    added += 1

p1.legend.location = "top_left"
p1.legend.click_policy = "hide"

p2 = figure(plot_width=800, plot_height=450)
# p2.title.text = 'Click on legend entries to hide the corresponding lines'

for i, (key, values) in enumerate(zero_ones):
    p2.line(range(len(values)), values, line_width=2, alpha=0.8, legend=key, color=Spectral11[i%11])

p2.legend.location = "top_right"
p2.legend.click_policy = "hide"

output_notebook()

show(gridplot([[p1], [p2]]))
```
